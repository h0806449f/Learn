常用函式

```python
import cv2
import numpy as np

cat= cv2.imread("cat.jpg")
cat= cv2.resize(cat, (0,0), fx=0.5, fy=0.5)   # resize 第二個參數: 輸出圖像大小 (0,0) 同原始大小

# 1 灰階
gray= cv2.cvtColor(cat, cv2.COLOR_BGR2GRAY)

# 2 模糊
blur= cv2.GaussianBlur(cat, (3,3), 0)
# 參數: (圖檔名稱, 模糊核的大小tuple, 標準差)
# 模糊核 的大小 須為基數   核越大，越模糊
# 標準差: 控制著高斯分布的形狀，標準差越大，高斯分布越平坦，圖像的模糊程度也就越大。0 自動最佳化

# 3 取圖形邊緣 (輪廓)
canny= cv2.Canny(cat, 20, 60)   
# 參數: (圖檔名稱, 最低門檻值, 最高門檻值)
# canny 如何判斷邊緣，圖形中A物體的邊緣 vs B物體的邊緣，像素差異很大
# 上下閥值高，被過濾掉的東西就會比較多

# 4 膨脹
kernel= np.ones((3, 3), np.uint8)
# 創立陣列 (結構元素，此結構大小影響膨脹的效果)
dilate= cv2.dilate(canny, kernel, iterations=1)
# 參數: (取得邊界的圖片, 結構元素, 跌代次數)
# 跌代次數也會影響膨脹結果

# 5 侵蝕 (膨脹的相反)
kernel_1= np.ones((3, 3), np.uint8)
erode= cv2.erode(dilate, kernel_1, iterations=2)
# 參數: (取得邊界的圖片, 結構元素, 跌代次數)

cv2.imshow("original", cat)
cv2.imshow("erode", erode)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

畫形狀 & 寫字

```python
import cv2
import numpy as np
# import

img= np.zeros( (600, 600, 3), np.uint8 )
# 600*600 彩色黑底圖形

cv2.line(img, (0,0), (img.shape[1],img.shape[0]), (255,255,255), 2)
# 畫直線
# 參數: (畫的圖, 起點, 終點, 顏色, 粗細)
# shape[1] 寬度，shape[0]高度

cv2.rectangle(img, (0,0), (200,200), (0,255,0), -1)
# 畫矩陣
# 參數: (畫的圖, 左上起點, 右下終點, 顏色, 粗細(填滿: -1 or cv2.FILLED)))
# # 色號排序[Blue, Green, Red]

cv2.circle(img, (300,300), 75, (255,0,0), 2)
# 畫圓形
# 參數: (畫的圖, 圓心, 半徑, 顏色, 粗細)

cv2.putText(img, "Python PP", (100,500), cv2.FONT_HERSHEY_PLAIN, 1, (255,255,255), 2)
# 寫文字
# 參數: (寫的圖, 文字, 起始座標(文字左下為起點), 字體, 字體大小, 字體顏色, 字體粗細)
# putText 不支援中文


cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

偵測圖片的特定顏色

```python
import cv2
import numpy as np

def empty(v):
    pass

img= cv2.imread("vegetable.jpg")
img= cv2.resize(img, (0,0), fx=0.5, fy=0.5)

hsv= cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
# 思序1: 如果需要辨識顏色，需要先將 BGR 轉換成 HSV (H色調,S飽和,V亮度)

cv2.namedWindow("TrackBar")
cv2.resizeWindow("TrackBar", 640, 320)
# 思序2: 創建額外視窗並調整視窗大小

cv2.createTrackbar("Hue Min", "TrackBar", 0, 179, empty)
cv2.createTrackbar("Hue Max", "TrackBar", 179, 179, empty) # 色調最小值最大值
cv2.createTrackbar("Sat Min", "TrackBar", 0, 255, empty)
cv2.createTrackbar("Sat Max", "TrackBar", 255, 255, empty) # 飽和最小值最大值
cv2.createTrackbar("Val Min", "TrackBar", 0, 255, empty)
cv2.createTrackbar("Val Max", "TrackBar", 255, 255, empty) # 亮度最小值最大值
# 最後參數: 此控制條被調整時，我們需要呼叫的function，先暫時用空白函式
# 思序3: 篩選顏色，需要六個控制條

while True:
    h_min= cv2.getTrackbarPos("Hue Min", "TrackBar")
    h_max= cv2.getTrackbarPos("Hue Max", "TrackBar")
    s_min= cv2.getTrackbarPos("Sat Min", "TrackBar")
    s_max= cv2.getTrackbarPos("Sat Max", "TrackBar")
    v_min= cv2.getTrackbarPos("Val Min", "TrackBar")
    v_max= cv2.getTrackbarPos("Val Max", "TrackBar")
    print(h_min, h_max, s_min, s_max, v_min, v_max)
    # getTrackbarPos 可以取得控制條上的值，第一參數:值, 第二參數:視窗名稱
    # 思序4: 輸出六個閥值


    lower= np.array([h_min, s_min, v_min]) # 最小值
    upper= np.array([h_max, s_max, v_max]) # 最大值

    selected_hsv= cv2.inRange(hsv, lower, upper)
    # inRange 第一參數:欲處理圖片,第二參數:最小值,第三參數:最大值 (min max 都要用array表示)
    # 思序5: 過濾顏色使用inRange


    result= cv2.bitwise_and(img, img, mask= selected_hsv)
    # bitwise_and 第一二參數:原圖片,第三參數:過濾後的黑白圖片
    # 暫時理解: 將img圖片，套上selected_hsv 這個遮罩，可以將我們要的顏色過濾出來
    # 思序6: 過濾出我們真正要的顏色
    
    
    cv2.imshow("img", img)
    cv2.imshow("selected_hsv", selected_hsv) # 篩選後，白色部分是我們要的區域
    cv2.imshow("result", result)
    cv2.waitKey(1)
    # 思序4: 在六個控制條上，即時反映數值

cv2.waitKey(0)
cv2.destroyAllWindows()
```

輪廓檢測 & 形狀檢測

```python
# 複合多種功能，方便複習
import cv2

img= cv2.imread("square.png")
# 讀取圖片
imgContour= img.copy()
# 先複製一張圖，畫輪廓用

cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# 轉灰階圖 因檢測輪廓不需要顏色

canny= cv2.Canny(img, 150, 200)
# 檢測邊緣 後續會利用此邊緣檢測輪廓 (圖,門檻值,門檻值)

contours, hierarchy= cv2.findContours(canny, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
# 檢測輪廓 (圖片, 模式, 方法)
# 模式: 外輪廓 內輪廓 內外都要 cv2.RETR_EXTERNAL 內外都要:cv2.RETR_TREE
# 方法: cv2.CHAIN_APPROX_NONE 儲存所有輪廓點
# 此函式會回傳兩個值 1 輪廓 2 階層 (階層暫時用不到)


for cnt in contours:
    # print(cnt)
    # 利用for迴圈，印出所有輪廓
    cv2.drawContours(imgContour, cnt, -1, (0,0,255), 4)
    # 畫出輪廓
    # 一參數 決定要畫在哪張圖上 (額外創一張圖img1)
    # 二參數 要畫的輪廓
    # 三參數 要畫哪一個輪廓 -1則全畫
    # 四參數 輪廓顏色
    # 五參數 輪廓粗細

    # print("體積:", cv2.contourArea(cnt))
    # 印出所有輪廓體積
    # 為了避免圖片上有雜訊，去除圖片上體積過小的圖案
    area= cv2.contourArea(cnt)
    if area > 500:

        print("邊長:", cv2.arcLength(cnt, True))
        # 計算輪廓邊長
        # 一參數 輪廓
        # 二參數 輪廓是否閉合

        peri= cv2.arcLength(cnt, True)
        # 取多邊形的邊
        vertices= cv2.approxPolyDP(cnt, peri*0.02, True)
        print("近似處理後的頂點數:", len(vertices), f"({len(vertices)}邊形)")
        # 對多邊形 近似處理 (目的:提高處理效率)
        # 一參數 多邊形的輪廓
        # 二參數 誤差值
        # 三餐數 多邊形是否閉合
        
        # 以下想用矩形，將圖片框出來
        x,y,w,h= cv2.boundingRect(vertices)
        cv2.rectangle(imgContour, (x,y), (x+w,y+h), (0,255,0), 4)

        # 在圖上標記 n邊形
        if len(vertices) == 6:
            cv2.putText(imgContour,"six",(x,y-5), cv2.FONT_HERSHEY_COMPLEX_SMALL,1,(0,0,0),1)
            # 一 欲處理圖 / 二 字 / 三 座標 / 四 字體 / 五 字體大小 / 六 顏色 / 七 字體粗細
        elif len(vertices) == 5:
            cv2.putText(imgContour,"five",(x,y-5), cv2.FONT_HERSHEY_COMPLEX_SMALL,1,(0,0,0),1)
        elif len(vertices) == 3:
            cv2.putText(imgContour,"three",(x,y-5), cv2.FONT_HERSHEY_COMPLEX_SMALL,1,(0,0,0),1)
        elif len(vertices) == 4:
            cv2.putText(imgContour,"four",(x,y-5), cv2.FONT_HERSHEY_COMPLEX_SMALL,1,(0,0,0),1)


cv2.imshow("img", img)
cv2.imshow("canny", canny)
cv2.imshow("imgContour", imgContour)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

人臉辨識  
官方提供的訓練文件: https://github.com/opencv/opencv/tree/4.x/data/haarcascades

```python
import cv2

img= cv2.imread("face.jpg")
gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# 轉灰階

faceCascade= cv2.CascadeClassifier("face_detect.xml")
# 載入我們下載的官方人臉辨識文件

faceRect= faceCascade.detectMultiScale(gray, 1.3, 5)
# 此函式會一個框框，一個框框檢驗有無人臉，檢測完後，會依照我們設定的倍數去縮小圖片
# 縮小後，會再一個框框，一個框框檢驗有無人臉，一直循環
# 一參數 欲處理圖片
# 二參數 圖片縮小倍數 (數值越小越嚴謹)
# 三參數 相鄰的框框，最少要有幾次 (圖中的臉，最少要被框到幾次，才正式算人臉 數值越大越嚴謹)

# 此函式會回傳矩形框框，因為有可能有很多人臉 會以list呈現

for (x,y,w,h) in faceRect:
# for in list
    cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0), 2)


cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
