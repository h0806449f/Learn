爬取 tabelog 網頁資料，並利用 pandas 將其儲存成 csv 檔案形式

```python
# tabelog : https://tabelog.com/tw/tokyo/rstLst/?SrtT=rt
from urllib.request import urlopen
from bs4 import BeautifulSoup
from urllib.error import HTTPError
import pandas as pd

df = pd.DataFrame(columns=["jp_title", "en_title", "rating", "d_price", "n_price", "url"])

slide = 55

while True:
    url = f"https://tabelog.com/tw/tokyo/rstLst/{slide}/?SrtT=rt"
    try:
        response = urlopen(url)
    except HTTPError:
        print("已經到最後一頁了")
        break

    page = BeautifulSoup(response)

    boxs = page.find_all("li", class_="list-rst")

    print(f"==========現在正在處理{slide}頁==========")
    for box in boxs:
        jp_title = box.find("small", class_ = "list-rst__name-ja")
        en_title = box.find("a", class_ = "list-rst__name-main")
        rating = box.find("b",class_ = "c-rating__val")
        n_price = box.find_all("span",class_ = "c-rating__val")[0]
        d_price = box.find_all("span",class_ = "c-rating__val")[1]
        connect = en_title["href"]
        # print(jp_title.text,en_title.text,rating.text)
        # print(connect)
        # print(d_price.text,n_price.text)

        # 將資料帶入pandas !!   這邊不熟!!
        # 思考1 將資料帶入series
        s = pd.Series(
            [jp_title.text,en_title.text,rating.text,d_price.text,n_price.text,connect],
            index = ["jp_title", "en_title", "rating", "d_price", "n_price", "url"]
        )

        # 思考2 將 series 帶入 dataframe
        # 思考3 是否忽略原本 index，創立新的索引值
        # 思考4 append 之後是一個新的dataframe 要取代回原本的df
        df = df.append(s, ignore_index=True)
    slide = slide + 1

# 思考5 儲存為csv檔案，encoding，index: 要不要儲存df自己建立的index? 選False
df.to_csv(r"練習1.csv",encoding="UTF-8",index = False)
```
