```python
# 作業練習
import pymysql
import prettytable
# import

membertable= prettytable.PrettyTable(["編號", "姓名", "生日", "地址"], encoding= "utf-8")
membertable.align["編號"]= "l"
membertable.align["姓名"]= "l"
membertable.align["生日"]= "l"
membertable.align["地址"]= "l"
# 空table

link= pymysql.connect(
    host= "0.tcp.jp.ngrok.io",
    user= "root",
    passwd= "",
    db= "2023-03-08",
    charset= "utf8",
    port= 18826)
# info from ngrok 

while True:
    print("(0) 離開程式")
    print("(1) 顯示會員列表")
    print("(2) 新增會員資料")
    print("(3) 更新會員資料")
    print("(4) 刪除會員資料")
    command= input("指令: ")
    # 迴圈提示指令

    if command== "1":
        cur= link.cursor()
        cur.execute("SELECT * FROM member")
        dataall= cur.fetchall()
        membertable.clear_rows()
        for d in dataall:
            membertable.add_row(d)
        print(membertable)

    elif command== "2":
        cur= link.cursor()
        data= {
        "a":input("新增_輸入姓名: "),
        "b":input("新增_輸入生日: "),
        "c":input("新增_輸入地址: ")}
        cur.execute("INSERT INTO `member` (name, birthday, address) VALUES (%(a)s,%(b)s,%(c)s)", data)
        link.commit()   

    elif command== "3":
        cur= link.cursor()
        data= {
        "a":input("更新_輸入姓名: "),
        "b":input("更新_輸入生日: "),
        "c":input("更新_輸入地址: ")}
        cur.execute("UPDATE `member` (name, birthday, address) VALUES (%(a)s,%(b)s,%(c)s)", data)
        link.commit()   
    
    elif command== "4":
        cur= link.cursor()
        data= {
        "a":input("刪除_輸入ID: ")}
        cur.execute("DELETE FROM `member` WHERE id=(%(a)s)", data)
        link.commit()   
    
    elif command== "0":
        link.close()
        break
```
